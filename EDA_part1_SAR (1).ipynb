```python
# @title
# box plot function
def box_p(n):
  sns.set_style("white")
  sns.set_style("ticks")
  plt.figure(figsize=(10, 5))
  sns.boxplot(data=data, x=data[n], color="lightblue", width=0.2);
  plt.title(f"Boxplot of the {n} of players")
  sns.despine(left=True);
```
